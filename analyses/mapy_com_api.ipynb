# Explore mapy.com API for raw address conversion

Mapy.com offers a geocoding API that can be used to convert raw address strings into structured address components.
Their API documentation can be found at: https://developer.mapy.com/rest-api-mapy-cz/function/geocoding/ and https://api.mapy.com/v1/docs/geocode/#/geocoding/api_geocode_v1_geocode_get.
API is free for first 200,000 requests, then it requires paid plan.
Let's explore the API with some example requests.

```python
import os

import requests
from dotenv import load_dotenv

load_dotenv("../configs/.env")

API_KEY = os.getenv("MAPY_COM_API_KEY")

response = requests.get(
    "https://api.mapy.com/v1/geocode",
    params={
        "query": "Reichstag berlin",
        "lang": "en",
        "limit": 5,
        "type": "regional.address",
        "apikey": API_KEY,
    },
)

response.json()
```

```python
items = response.json()["items"]
items
```

Response contains all information we need for geojson conversion. We can extract the "items" field from the response JSON to get the list of geocoded items.
Then select the first item and get all its address components.
