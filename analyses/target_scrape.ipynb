Germany insolvency portal has no API access. Only way to get data is to scrape the web interface. The website uses java script so need to use selenium. The problem with selenium is that it is slow. Another problem with insolvency portal is that the search results dont contain address of the person/company in insolvency. Only way to get address is to click into detail page of each search result, where the address is present. Unfortunately address is not easily separable from other text. So Im not sure if insolvency portal of germany is even scrapeable.

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait

driver = webdriver.Chrome()
driver.get("https://neu.insolvenzbekanntmachungen.de/ap/suche.jsf")

# Wait for form to load (JavaScript might build it dynamically)
wait = WebDriverWait(driver, 10)

# Fill name field
name_field = wait.until(EC.presence_of_element_located((By.ID, "frm_suche:litx_firmaNachName:text")))
name_field.send_keys("sm*")

# Submit form
submit_button = driver.find_element(By.ID, "frm_suche:cbt_suchen")
driver.execute_script("arguments[0].click();", submit_button)

# Wait for response/redirect
wait.until(EC.url_changes(driver.current_url))
```
